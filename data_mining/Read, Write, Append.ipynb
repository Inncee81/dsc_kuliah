```python
import PyPDF2 
import textract
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

Read text from pdf

```python
pdfFileObj = open('Ayah.pdf', 'rb')
reader = PyPDF2.PdfFileReader(pdfFileObj)
num_pages = reader.numPages
count = 0
text = ""
while count < num_pages:
    pageObj = reader.getPage(count)
    count +=1
    text += pageObj.extractText()
text.encode("utf-8")
print(pageObj.extractText())
```

Create new file to save the text

```python
pdfFileObj2 = open("ayah.txt", "x")
```

Append text to the new file

```python
pdfFileObj2 = open("ayah.txt", "a", encoding='utf-8')
pdfFileObj2.write(text)
pdfFileObj2.close()
pdfFileObj2 = open("ayah.txt", "r", encoding='utf-8')
print(pdfFileObj2.read())
```
