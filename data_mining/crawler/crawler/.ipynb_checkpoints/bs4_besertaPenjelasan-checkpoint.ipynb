```python
from bs4 import BeautifulSoup
import requests
import json 
import signal
```

# Functions!

## The loop function

```python
def theLoop(queue, D, E, title, seed):
    for url in queue.copy():
        queue.remove(url)
        page1 = requests.get(url)
        soup = BeautifulSoup(page1.text, "html.parser")
#         print(url)
        
        D[url] = 'None' if soup.title is None else soup.title.string
        links = soup.find_all("a")
        old_title = title
        title = 'None' if soup.title is None else soup.title.string
        for a in links:
            E.append(old_title+": "+title)
            if a.has_attr('href') and (a['href'].startswith(seed)) and (a['href'] not in D.values()) and (a['href'] not in queue):
                queue.add(a['href'])
        print(len(D))
    if len(queue.difference(set(D.keys()))) !=0 :
#         print("Queue: ", queue)
        new_queue = queue.difference(set(D.keys()))
        page1 = requests.get(list(new_queue)[0])
        soup = BeautifulSoup(page1.text, "html.parser")
        links = soup.find_all("a")
        queue = set()
        title = 'None' if soup.title is None else soup.title.string
        D, E = theLoop(new_queue, D, E, title, seed)
    else:
        return D, E
    return D, E
```

## Store D and E to a text/json file
Function ini untuk menyimpan hasil crawling ke dalam sebuah file txt
* file D: kumpulan link dan judul, berupa dictionary (key: url, value: judul halaman) --> dibalik karena ada beberapa page yang memiliki title sama
* file E: kumpulan mapping (asal: tujuan)

```python
def store_results(D, E, filename):
    fileD = open("D_"+filename+".txt", "x")
    fileD.write(json.dumps(D))
    fileD.close()
    fileE = open("E_"+filename+".txt", "x")
    fileE.write(json.dumps(E))
    fileE.close()
```

# The "Simple" Crawler
Mengikuti algoritma Simple Crawler yang dibahas di kelas
* Seed URL berupa homepage, diberikan pada saat kita memanggil function crawler
* Function 'theLoop' adalah function recursive

```python
def crawler(seed, filename):
    page1 = requests.get(seed)
    soup = BeautifulSoup(page1.text, "html.parser")
    links = soup.find_all("a")
    queue = set()
    title = 'None' if soup.title is None else soup.title.string
    D = {seed : title}
    E = []
    for a in links:
        if a.has_attr('href') and (seed in a['href']):
            queue.add(a['href'])
    D, E = theLoop(queue, D, E, title, seed)
#     print(D.keys())
#     print(E)
    store_results(D, E, filename)
```

## Crawler: until layer ke N
Hanya sedikit memodifikasi basic simple crawler:
* memodifikasi bagian pemanggilan function loop
* function 'theLoop_n' sedikit berbeda dari 'theLoop': 'theLoop_n' bukan function recursive

```python
def crawler_n(seed, filename):
    page1 = requests.get(seed)
    soup = BeautifulSoup(page1.text, "html.parser")
    links = soup.find_all("a")
    queue = set()
    title = 'None' if soup.title is None else soup.title.string
    D = {seed : title}
    E = []
#     layer = int(input("Depth: "))
    layer = 3
    count = 1
    for a in links:
        if a.has_attr('href') and (seed in a['href']):
            queue.add(a['href'])
#     print(len(D))
    while layer>count:
        if len(queue.difference(set(D.keys()))) !=0 :
#         print("Queue: ", queue)
            new_queue = queue.difference(set(D.keys()))
            page1 = requests.get(list(new_queue)[0])
            soup = BeautifulSoup(page1.text, "html.parser")
            links = soup.find_all("a")
            D, E, queue = theLoop_n(queue, D, E, title, seed)
        count += 1
#     print(D.keys())
#     print(E)
    store_results(D, E, filename)
```

```python
def theLoop_n(queue, D, E, title, seed):
    for url in queue.copy():
        queue.remove(url)
        page1 = requests.get(url)
        soup = BeautifulSoup(page1.text, "html.parser")
#         print(url)
        D[url] = 'None' if soup.title is None else soup.title.string
        links = soup.find_all("a")
        old_title = title
        title = 'None' if soup.title is None else soup.title.string
        for a in links:
#             E.append(old_title+": "+title)
            if a.has_attr('href') and (a['href'].startswith(seed)) and (a['href'] not in D.values()) and (a['href'] not in queue):
                E.append(url+": "+a['href'])
                queue.add(a['href'])
        print(len(D))
    return D, E, queue
```
