# PREPROCESING
DISINI KITA MENGHILANGKAN SPASI YANG TIDAK DIPAKAI LALU MENGECILKAN SEMUA HURUF BESAR SEHINGGA TIDAK ADA KATA YANG REDUDANT KETIKA DI LAKUKAN PERHITUNGAN VEKTOR

CONTOH ayub dengan Ayub nanti kalau tidak di kecilkan maka bisa dianggap sebagai sebuah kata yang berbeda

```python
df['konten']=df['konten'].str.strip()
```

```python
df['konten']=df['konten'].str.replace('"',"")
```

```python
df['konten']=df['konten'].str.lower()
```

# MELAKUKAN VEKTORISASI DENGAN TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

KITA MEMBATASI HANYA 100 KATA DENGAN NILAI PALING TINGGI dan dengan nilai bobot lebih dari 0.05 dan dibawah 0.9

```python
tf=TfidfVectorizer(max_features=100,min_df=0.05,max_df=0.9)
```

```python
hasil=tf.fit_transform(df.konten)
tabel=pd.DataFrame(hasil.toarray(),columns=tf.get_feature_names())
```

BERIKUT DIBAWAH ADALAH HASIL DARI TF-IDF YANG SUDAH DINORMALISASI DENGAN L2

```python
tabel
```

```python
tabel.sum().sort_values(ascending=False).head(20)
```

karena didapat bahwa ada banyak kata hubung maka kita akan buat ulang yang tidak ada kata hubungnya DENGAN MENGGUNAKAN PARAMETER STOP WORDS KITA DAPAT TIDAK MENGIKUT SERTAKAN KATA HUBUNG DALAM PERHITUNGAN

```python
tf=TfidfVectorizer(max_features=100,min_df=0.05,max_df=0.9,stop_words=['dan','itu','yang','di','ke','atau','akan','dengan','serta','atau','tapi','namun','sebab','karena','kalau','untuk','guna','supaya','hingga'])
hasil=tf.fit_transform(df.konten)
tabel=pd.DataFrame(hasil.toarray(),columns=tf.get_feature_names())
tabel.sum().sort_values(ascending=False).head(25)
```

BERIKUT ADALAH TOP 20 WORD DI KITAB KEJADIAN

# MELAKUKAN CLUSTERING
DENGAN MENGGUNAKAN ALGORITMA KMEANS KITA AKAN MENCARI TITIP CLUSTER OPTIMAL DENGAN MENGGUNAKAN ELBOW METHOD

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
wcss = []
for i in range (1,20):
    km = KMeans(n_clusters= i,
                 # max iteration by run
                random_state=1)
    km.fit(tabel)
    wcss.append(km.inertia_)  # inertia = within-cluster sum-of-squares (WCSS) 
plt.figure(figsize=(10,10))
plt.plot(range(1,20), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

DIDAPATI BAHWA SEKITAR 8 CLUSTER ADALAH TITIK OPTIMUM

```python
km=KMeans(n_clusters=8,random_state=1)
km.fit(tabel)
hasil=km.predict(tabel)
```

setelah mendapat hasil cluster maka kita sisipkan dikolom target

```python
tabel2=tabel.copy()
tabel2['target']=pd.DataFrame(hasil)
tabel2
```

# TOP 10 KEYWORD SETIAP CLUSTER

```python
def get_top_keywords(data, clusters, labels, n_terms):
    df = pd.DataFrame(data).groupby(clusters).mean()
    
    for i,r in df.iterrows():
        print('\nCluster {}'.format(i+1))
        print(','.join([labels[t] for t in np.argsort(r)[-n_terms:]]))
            
get_top_keywords(tabel, tabel2.target, tf.get_feature_names(), 10)
```

# LAKUKAN VISUALISASI

```python
plt.figure(figsize=(10,8))
sns.scatterplot(model[:,0],model[:,1],hue=tabel2.target,palette="coolwarm")
plt.show()
```
