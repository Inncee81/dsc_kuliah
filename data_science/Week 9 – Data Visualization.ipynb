# Import Necessary Library

```python
# Import library untuk visualisasi data
import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd

#plt.style.use("fivethirtyeight")
```

Membuat Line Plot sederhana

```python
bulan = [0, 2, 4, 6, 8, 10, 12]
omzet_miliar = [1, 3, 4, 7, 2, 8, 9]

plt.plot(bulan, omzet_miliar)
plt.title("Omzet Penjualan 2019")
plt.xlabel("Bulan")
plt.ylabel("dalam Miliar Rupiah")
plt.show()
```

Membuat Pie Plot

```python
items = ["Product 1", "Product 2", "Product 3", "Product 4", "Product 5"]
proportions = [40, 15, 30, 50, 25]
pie_explode = [0, 0.2, 0, 0, 0]

plt.pie(proportions, labels=items, startangle=45, shadow=True, explode=pie_explode)
plt.axis("equal")
plt.title("List of Defects", pad=20)
plt.show()
```

```python
x_units = [1, 2, 3, 4]
y_units = [10, 24, 36, 48]
tick_label = ["Q1", "Q2", "Q3", "Q4"]

plt.bar(x_units, y_units, tick_label=tick_label,
        width=0.8, color=["red", "black"])

plt.xlabel("2019 Profit Report")
plt.ylabel("in Billion USD")
plt.title("Company Profit", pad=20)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure()

languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# change the bar colors to be less bright blue
bars = plt.bar(pos, popularity, align='center', linewidth=0, color='lightslategrey')

# make one bar, the python bar, a contrasting color
bars[0].set_color('#1F77B4')

# soften all labels by turning grey
plt.xticks(pos, languages, alpha=0.8)
plt.ylabel('% Popularity', alpha=0.8)
plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# remove all the ticks (both axes), and tick labels on the Y axis
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)
plt.show()
```

# Menggunakan Seaborn

```python
# Read dataset
file_name = "https://raw.githubusercontent.com/twiradinata/ISB2017-applied-data-science/master/pokemon_collection.csv"
df = pd.read_csv(file_name, delimiter=";", index_col = 0)
df.head()
```

```python
# Displaying Scatter Plot
sns.lmplot(x='Attack', y='Defense', data=df)
```

```python
# Setting Scatterplot's arguments
sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False, # Without regression line
           hue='Stage')   # Give color by evolution stage
```

```python
# Boxplot menampilkan sebaran data kita
# Yang didalam Box adalah Q1 s.d. Q3

sns.boxplot(data=df)
```

```python
# Data "Total" "Stage" dan "Legendary" tidak diperlukan untuk Individual Combat Stats
# Kita buat DataFrame khusus untuk Individual Combat Stats

indiv_combat_stats_df = df.drop(["Total", "Stage", "Legendary"], axis=1)

# Boxplot yang baru
sns.boxplot(data=indiv_combat_stats_df)
```

```python
# Hitung korelasi antar variabel
corr = indiv_combat_stats_df.corr()
 
# Gambarkan kekuatan korelasi dengan Heatmap
sns.heatmap(corr, vmin=-1, vmax=1)
```

```python
# Distribution Plot (Histogram) untuk Speed
sns.distplot(indiv_combat_stats_df.Speed)
```

```python
# Drop data yang bersifat agregasi (Total) dan Ketegorical (Legendary)
indiv_df = df.drop(["Total", "Legendary"], axis=1)

# Pair plot menampilkan scatter plot antar 2 variable untuk seluruh variabel numerik
# Gunakan variable "Stage" untuk membedakan warna plot
sns.pairplot(data=indiv_df, hue="Stage")
```

# Exercise dengan Iris Dataset

Dataset bunga irisadalah kumpulan data yang diperkenalkan oleh ahli statistik Inggris, Ronald Fisher. Dataset ini terdiri dari 50 sampel dari masing-masing tiga spesies Iris (Iris setosa, Iris virginica dan Iris versicolor). Empat fitur diukur dari setiap sampel: panjang-lebar sepal dan petal, dalam sentimeter. Berdasarkan kombinasi keempat ciri ini, Fisher mengembangkan model diskriminan linier untuk membedakan spesies satu sama lain.<br>
<img alt="Parts of Flower" src="https://cdn.britannica.com/s:700x500/39/91239-004-44353E32/Diagram-flowering-plant.jpg">

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Gunakan data Iris untuk membuat box plot
df = sns.load_dataset('iris')
df.head()
```

Buatlah Box Plot untuk melihat sebaran data pada kolom "sepal_length", setting lebar box plot 0.4

```python
df['sepal_length']
```

```python
sns.boxplot(data=df['sepal_length'], width=0.4)
```

Buatlah Box Plot untuk melihat sebaran data pada kolom "sepal_length", setting lebar box plot 0.4. Khusus untuk spesies "virginica"

```python
df[df['species'] == 'virginica']['sepal_length']
```

```python
data1 = df.loc[df['species'] == 'virginica', 'sepal_length']
sns.boxplot(data1, width=0.4)
```

<h4>Boxplot Horizontal</h4>
Boxplot bisa berupa plot horizontal. Contoh di bawah ini menunjukkan diagram kotak dalam arah horizontal. Kami menggunakan dataset iris lagi. Warna yang ditampilkan adalah warna default, tetapi dapat dikonfigurasi.

Buatlah Boxplot horizontal untuk menampilkan "sepal_width" berdasarkan ketiga jenis species.

```python
 sns.boxplot(x=df['sepal_width'], y=df['species'], data=df['sepal_width'], width=0.4)
```

<h4>Warna pada Boxplot</h4>
Warna plot warna boxplot dapat dikonfigurasi. Anda dapat melakukannya dengan meneruskan parameter palet. Ada berbagai macam palet, misalnya palet "blues" adalah semua jenis biru.
<p>Daftar palet seaborn dapat dilihat <a href="https://medium.com/@morganjonesartist/color-guide-to-seaborn-palettes-da849406d44f">disini</a>

Buatlah tiga box plot yang menggambarkan sebaran data pada kolom "sepal_width" untuk masing-masing jenis species. Gunakan palet "coolwarm" dan setting lebar box plot pada 0.5. Berilah judul "Distribusi Sepal Width untuk ketiga kategori spesied" untuk box plot ini.

```python
sns.boxplot(y=df['sepal_width'], x=df['species'], data=df['sepal_width'], width=0.4, palette='coolwarm').set_title('Distribusi Sepal Width')
```

Buatlah histogram plot untuk menampilkan frekuensi sepal_length khusus untuk spesies "versicolor". Beri judul plot "Frekuensi distribusi sepal length untuk spesies versicolor" dengan padding 20.

```python
plt.hist(df['sepal_length'], bins=10)
plt.title('Frekuensi distribusi sepal length untuk spesies versicolor', pad=20)
plt.show()
```

Buatlah heatmap yang menunjukkan kekuatan korelasi antara seluruh variabel numerik khusus spesies versicolor. Gunakan palet warna "coolwarm", title "Versicolor: Korelasi antar variabel", atur skala korelasi dari -1 hingga 1.

```python
sns.heatmap(df[1:].corr(), cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Versicolor: Korelasi antar variabel', pad=20)
```

Buatlah Pair Plot untuk menampilkan hubungan seluruh variabel numerik pada dataset Iris ini. Gunakan warna palet "BrBG" dan variabel "species" untuk membedakan warna.

```python
sns.pairplot(data=df, palette='BrBG', hue='species')
```
