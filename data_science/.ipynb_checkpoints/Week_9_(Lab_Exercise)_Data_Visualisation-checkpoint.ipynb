# 3. Kategorikan "Age" menjadi 3 Kategori (Muda, Dewasa, Usia Lanjut) diberi nama kolom baru "Age Category" dengan menggunakan Quantile [0, 0.25, .75, .9]

```python
# 3.1 Buatlah sebuah fungsi categorize_age yang akan di implementasikan pada Series Age
df['Age'].quantile([0, 0.25, 0.75, 0.9])
```

```python
def categorize_age(age):
    if age < 35:
        return "Muda"
    elif age >= 35 and age <= 47:
        return "Dewasa"
    elif age > 47:
        return "Usia Lanjut"
    else:
        return None
```

```python
# 3.2 Implmentasikan categorize_age ke Series Age disini
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Age'].apply(categorize_age)
```

```python
# 3.3 Visualisasikan Hubungan antara  "Age Category" dengan "Survived"
df['Age_category'] = df['Age'].apply(categorize_age)
sns.countplot(x='Age_category', data=df, hue='Survived')
plt.show()
```
