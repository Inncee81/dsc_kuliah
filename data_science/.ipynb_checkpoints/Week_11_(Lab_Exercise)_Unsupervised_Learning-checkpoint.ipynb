```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering, KMeans
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
df.head()
```

```python
# Lakukan Eksplorasi Data dan K Mean Clustering dan Hierarchical pada dataset iris
```

### K-Means Clustering

```python
# Label encoding
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])

# Scaling Data
sc = StandardScaler()
dfs = StandardScaler().fit_transform(df)

# Scatter Plot Elbow
wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters = i,
               n_init = 3,
               max_iter = 500,
               random_state = 42)
    km.fit(dfs)
    wcss.append(km.inertia_)
    
plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show
```

```python
km_model = KMeans(n_clusters = 3, n_init = 10, max_iter = 500, random_state=42)
y_kmeans = km_model.fit_predict(dfs)
df['K-Means Cluster'] = y_kmeans
df.groupby(["K-Means Cluster", "species"]).agg(SepalL = ('sepal_length', 'mean'),
                                              SepalW = ('sepal_width', 'mean'),
                                              PetalL = ('petal_length', 'mean'),
                                              PetalW = ('petal_width', 'mean'))
```

### Hierarchical Clustering

```python
# Label encoding
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])

# Scaling Data
sc = StandardScaler()
dfs = StandardScaler().fit_transform(df)

# Dendrogram
dist_matrix = sch.linkage(dfs, method='complete', metric='euclidean')
# dist_matrix = sch.linkage(dfs, method='average', metric='euclidean')
# dist_matrix = sch.linkage(dfs, method='weighted', metric='euclidean')
# dist_matrix = sch.linkage(dfs, method='ward', metric='euclidean')
# dist_matrix = sch.linkage(dfs, method='median', metric='euclidean')
# dist_matrix = sch.linkage(dfs, method='centroid', metric='euclidean')
# dist_matrix = sch.linkage(dfs, method='single', metric='euclidean')
dn = sch.dendrogram(dist_matrix)
plt.title("Dendrogram", pad=20)
plt.xlabel("Iris")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage='ward')
y_hc = hc.fit_predict(dfs)

df['Hierarchical Cluster'] = sch.fcluster(dist_matrix, 3, criterion='maxclust')
df.groupby(['Hierarchical Cluster', 'species']).agg(SepalL = ('sepal_length', 'mean'),
                                                    SepalW = ('sepal_width', 'mean'),
                                                    PetalL = ('petal_length', 'mean'),
                                                    PetalW = ('petal_width', 'mean'))
```
