# Clustering

## K-Means Clustering

### Import Library yang Diperlukan

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

### Load Dataset

````python
url = "https://raw.githubusercontent.com/twiradinata/ISB2017-applied-data-science/master/mall_customers.csv"
df = pd.read_csv(url)
df.head()
````

````python
# Check apakah sdh tidak ada missing data (Non-Null Count)
df.info()
````

### Data Preparation
Drop kolom yang tidak dapat digunakan untuk clustering

````python
df.drop(["CustomerID"], axis=1, inplace=True)
````

### Label Encoding
Data categorical diubah menjadi angka

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"]) #0 female, 1 Male
````

````python
df.head()
````

### Feature Scaling

````python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
dfs = StandardScaler().fit_transform(df)
````

### Plot untuk Estimasi Clustering

````python
sns.pairplot(df, hue="Gender")
````

### K-Means Algorithm
How it works? Basically:

    1- initial clusters centers (centroids) are randomly chosen
    2- the observations are assigned to the closest centroid, based in some distance measure
    3- the centroids are recalculated with the means of the observations that makes part of each cluster.

Process 2 and 3 happen repeatedly, with the goal of minimize the total within cluster variation, until have no changes or reach some tolerance.

With K-means we have to pre-specify the number of clusters. Here, we gonna use the elbow method, that is, we gonna try some numbers of clusters (n) and observe the within clusters variation related to each n. When the sum of a cluster does not mean a considerable reduction of the within clusters variation we have a possible good choice of n (elbow on the graph).

````python
from sklearn.cluster import KMeans

wcss = []
for i in range (1,11):
    km = KMeans(n_clusters= i,
                n_init=5,  # run 5 times with different random inicial centroids
                max_iter=500,  # max iteration by run
                random_state=1)
    km.fit(dfs)
    wcss.append(km.inertia_)  # inertia = within-cluster sum-of-squares (WCSS) 

plt.plot(range(1,11), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
````

### Choose number of Clusters
Training model K-Means pada dataset dengan jumlah cluster 5

````python
km_model = KMeans(n_clusters=5, n_init=5, max_iter= 500, random_state=1)

y_kmeans = km_model.fit_predict(dfs)
````

### Conclude Clusterization Result

````python
df["K-Means Cluster"] = y_kmeans

df.groupby("K-Means Cluster").agg(Age = ('Age', 'mean'),
                                 AnnualIncome = ('Annual Income (k$)', 'mean'),
                                 SpendingScore=('Spending Score (1-100)', 'mean'),
                                 Gender=('Gender', 'mean'),
                                 Count=("K-Means Cluster", 'count'))
````

Hasil clustering menunjukkan karakteristik dari 5 group sbb:

* Grup dengan usia rata-rata 38thn, high income, spending rendah, pria/wanita.
* Grup dengan usia rata-rata 28thn, medium income, spending cukup tinggi, wanita.
* Grup dengan usia rata-rata 48thn, medium income, spending cukup rendah, wanita.
* Grup dengan usia rata-rata 56thn, medium income, spending cukup rendah, pria.
* Grup dengan usia rata-rata 28thn, medium income, spending cukup tinggi, pria.

## Hierarchical Clustering

Hierarchical Clustering adalah metode analisis kelompok yang berusaha untuk membangun sebuah hirarki kelompok data.

Strategi pengelompokannya umumnya ada 2 jenis yaitu Agglomerative (Bottom-Up) dan Devisive (Top-Down).

How it works? Basically in case of agglomerative with complete linkage:

```
*    1- a distance (similarity) matrix is calculated.
*    2- single samples (inicially considered as clusters) will be merged based on distance between the most dissimilar samples.
*    3- the matrix is updated
*    4- the steps are repeated until remains one cluster with all observations.
```




### Import Library yang Diperlukan

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

### Load Dataset

````python
url = "https://raw.githubusercontent.com/twiradinata/ISB2017-applied-data-science/master/mall_customers.csv"
df = pd.read_csv(url)
df.head()
````

### Data Preparation
Drop kolom yang tidak dapat digunakan untuk clustering

````python
df.drop(["CustomerID"], axis=1, inplace=True)
````

### Label Encoding

````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])
````

### Feature Scaling

````python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
dfs = StandardScaler().fit_transform(df)
````

### Plot untuk Estimasi Clustering
Using the dendrogram to find the optimal numbers of clusters.

````python
import scipy.cluster.hierarchy as sch

dist_matrix = sch.linkage(dfs, method='complete', metric='euclidean')
dn = sch.dendrogram(dist_matrix)
plt.title("Dendogram", pad=20)
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
````

### Hierarchical Clustering Algorithm

````python
from sklearn.cluster import AgglomerativeClustering 

hc = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage ='ward')
y_hc = hc.fit_predict(dfs)
````

### Conclude Clusterization Result

````python
print(df)
````

````python
df['Hierarchical Cluster'] = sch.fcluster(dist_matrix,4, criterion='maxclust')

df.groupby("Hierarchical Cluster").agg(Age = ('Age', 'mean'),
                                 AnnualIncome = ('Annual Income (k$)', 'mean'),
                                 SpendingScore=('Spending Score (1-100)', 'mean'),
                                 Gender=('Gender', 'mean'),
                                 Count=("Hierarchical Cluster", 'count'))
````

Hasil clustering menunjukkan karakteristik dari 4 group sbb:

* Kelompok pria/wanita, rata-rata usia 33thn dengan high income dan high spending score.
* Kelompok pria/wanita, rata-rata usia 40thn dengan high income dan very low spending score.
* Kelompok pria/wanita, rata-rata usia 55thn dengan low income dan low spending score.
* Kelompok pria/wanita, rata-rata usia 27thn dengan low income and medium spending score.
